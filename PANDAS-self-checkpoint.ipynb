## SERIES

```python
import numpy as np
import pandas as pd

ser = pd.Series([4, 7 ,-5, 3], index = ['a', 'b', 'c', 'd'])
print("*",ser)
print("1",ser.values)
print("2",ser.index)
print("3","b" in ser)#checking index in ser
print("4", "f" in ser)
print("5", ser<0)#values less than 0 returns answer in boolean
print("6", ser['a'])#value of single index
print("7", ser[['c', 'd']])#for more than 1 indexes we used two square brackets
```

```python
data = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
print("1\n",pd.Series(data))#incase of passing dictionary it will automaticallly sort its index
states = ['Ohio','Oregon','Texas','Utah']
print("2\n",pd.Series(data, index = states ))#incase of passing index of its own it will return values according to desired index
print("3\n",pd.Series(data, index = ['California', "Ohio","Texas", "Utah"]))#it will give NAN to california as it does not have any values
data1 = pd.Series(data, index = ['California', "Ohio","Texas", "Utah"])
print("4\n", pd.isnull(data1))#it will check NAN value in given series
print("5\n", pd.notnull(data1))
data1["California"] = 1000#it will change its previous value
display(data1)
```

## DATA FRAME

```python
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],        
        'year': [2000, 2001, 2002, 2001, 2002, 2003],        
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
display(pd.DataFrame(data))#this will make keys to index and values to values
data1 = pd.DataFrame(data, columns = ['year', 'state', 'pop'])#we can also assign columns of our desired
display(data1)
data2 = pd.DataFrame(data, columns = ['year', 'state', 'pop'], index = ['one', 'two', 'three', 'four', 'five', 'six'])
display(data2)
print(data1['state'])
#Or
data1.state#can be access through attribute
data3 = pd.DataFrame(data, columns = ['year', 'state', 'pop','dept'], index = ['one', 'two', 'three', 'four', 'five', 'six'])
display(data3)#as one unkown column debt has been enter so it will give NAN under it
ln = len(data3)#this gives the length of rows
print(ln)
#to assign the value of debtdata3 = pd.DataFrame(data, columns = ['year', 'state', 'pop','dept'], index = ['one', 'two', 'three', 'four', 'five', 'six'])
data4 = pd.DataFrame(data, columns = ['year', 'state', 'pop','dept'], index = ['one', 'two', 'three', 'four', 'five', 'six'])
rng = np.arange(ln)
data4['debt'] = rng
display(data4)
#OR we can also assign the value of column debt as;
data5 = pd.DataFrame(data, columns = ['year', 'state', 'pop','dept'], index = ['one', 'two', 'three', 'four', 'five', 'six'])
val = pd.Series([-1.2, -1.5, -1.7, 2, 6, 0.5], index=['one','two', 'three', 'four', 'five', 'six'])
data5['dept'] = val
display(data5)
#Or
data5['dept'] = val.values
display(data5)
print(data5['dept'][0])
```

## REINDEX

```python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
display(obj3)
#re-index will reindexes the given series and thus put NAN on un given values
display(obj3.reindex(range(9)))
#Now to get rid from NAN we used ffill which put previous values in place of NAN
display(obj3.reindex(range(9), method = 'ffill'))
states = pd.DataFrame(np.arange(9).reshape((3,3)), index=['a','c','d'], columns=['Ohio', 'Texas', 'California'])
print(states)
print()
states1 = states.reindex(['a', 'b', 'c', 'd'])#it gives NAN value at index b
print(states1)
print()
states2 = states.reindex(['a', 'b', 'c', 'd'], method='ffill')#use for rows, it will remove NAN
print(states2)
```
