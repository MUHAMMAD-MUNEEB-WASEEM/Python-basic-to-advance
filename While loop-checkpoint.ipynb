```python
a=1
while a<10:
    print(a,"a chota hai")
    a = a+1
```

```python
flag = True
while flag:
    name = input("What is your name sir...")
    print(f"Welcome to class sir {name}")
    if name == "muneeb":
        flag=False
```

```python
prompt = "\nPlease enter the name of a city you have visited:"
prompt += "\n(Enter 'quite' when you are finished.)"

while True:
    city = input(prompt)
                 
    if city == "quite":
        break
    else:
        print("I would love to go to "+city.title()+"!")
```
