### SERIES

```python
#A Series is a one-dimensional array-like object containing a sequence of values (of similar types to NumPy types) and an 
#associated array of data labels, called its index. 

obj = pd.Series([4, 5, 6, 7], index = ['a', 'b', 'c', 'd'])
display(obj)
############## values ################
print(obj.values)
############## index ##################
print(obj.index)
############ calling #################
print(obj['a'])
print(obj[['a','b','c']])
############ changing value ############
obj['d'] = 8
print(obj)
########### boolean #####################
print(obj>4)
print("*",'a' in obj)# works on index
print("*",'e' in obj)
################### actual Values #############
print(obj[obj>4])
################# mathematical ################
obj1 = obj*2
print(obj1)
################# change in whole index ############
obj1.index = ['Bob', 'Steve', 'Jeff', 'Ryan']
print(obj1)
################ numpy operations #################3
print(np.exp(obj))
############### dict to series ####################
#Should you have data contained in a Python dict, you can create a Series from it by passing the dict:
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
objd = pd.Series(sdata)
print(objd)
#When you are only passing a dict, the index in the resulting Series will have the dict’s keys in sorted order. 
#You can override this by passing the dict keys in the order you want them to appear in the resulting Series:
states = ['California', 'Ohio', 'Oregon', 'Texas']#no california in original dict so return NAN
objd1 = pd.Series(sdata, index = states)
print(objd1)
print('California' in objd1)# although its value is NAN but it has appeared as index in series
print('karachi' in objd1)
################# is null ###############
print(pd.isnull(objd1))
#OR
print(objd1.isnull())
################# not null ##############
print(pd.notnull(objd1))
#OR 
print(objd1.notnull())
objdR = objd + objd1#It will add value index wise and if any of index is not present in one of series than it gives NAN
print(objdR)
############## name of series ##############
objd1.name = 'population'
print(objd1)
############# name of index ################
objd1.index.name = 'states'
print(objd1)
```

### REINDEXING

```python
################## with series ####################3

obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
display(obj)
#Calling reindex on this Series rearranges the data according to the new index, introducing missing values 
#if any index values were not already present:
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e'])
display(obj2)
#################3 ffill ##################
#For ordered data like time series, it may be desirable to do some interpolation or filling of values when reindexing. 
#The method option allows us to do this, using a method such as ffill, which forward-fills the values
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
display(obj3)
obj4 = obj3.reindex(np.arange(6)) 
display(obj4)
obj5 = obj3.reindex(np.arange(6), method = 'ffill') 
display(obj5)

#################### with DataFrame ####################

frame = pd.DataFrame(np.arange(9).reshape((3, 3)),
                      index=['a', 'c', 'd'],
                      columns=['Ohio', 'Texas', 'California'])
display(frame)
frame2 = frame.reindex(['a', 'b', 'c', 'd'])
display(frame2)
#to get rid from NAN in index b we again use ffill
frame3 = frame.reindex(['a', 'b', 'c', 'd'], method = 'ffill')
display(frame3)
#The columns can be reindexed with the columns keyword:
display(frame3.reindex(columns = ['karachi', 'lahore', 'islamabad']))#it gives nan as no column present with such names
display(frame3.reindex(columns=['Texas', 'Utah', 'California']))
```

### SELECTION WITH LOC AND ILOC

```python
#For DataFrame label-indexing on the rows, I introduce the special indexing operators loc and iloc. 
#They enable you to select a subset of the rows and columns from a DataFrame with NumPy-like notation 
#using either axis labels (loc) or integers (iloc). 

#loc -> [row-label, column-label]
#iloc -> [row-no, column-no]

data = pd.DataFrame(np.arange(16).reshape((4, 4)),
                    index=['Ohio', 'Colorado', 'Utah', 'New York'],
                    columns=['one', 'two', 'three', 'four']
                   )
display(data)
display(data.loc['Colorado', ['two']])
display(data.loc['Colorado', ['two', 'three']])
display(data.iloc[1, [1,2]])
display(data.iloc[2, [3, 0, 1]])
display(data.iloc[2])#only row name specify so give row 2 (Utah) completely 
display(data.iloc[[1, 2], [3, 0, 1]])
display(data.loc[:'Utah', ['two']])
display(data.loc[:'Utah', ['two','three']])#label end inclusive
display(data.iloc[:2, [1,2]])#number end exclusive
display(data.iloc[:2, :3])#number end exclusive
display(data.iloc[:, :3][data.three > 6])
```

### Filling In Missing Data

```python
from numpy import nan as NA
df = pd.DataFrame(np.random.randn(7, 3))
display(df)
df1 = df
df1.iloc[:4, 1] = NA
df1.iloc[:2, 2] = NA
display(df1)
#the fillna method is the workhorse function to use. Calling fillna with a constant replaces missing values with that value
display(df1.fillna(0))
#Calling fillna with a dict, you can use a different fill value for each column:
display(df1.fillna({1: 0.5, 2:0.1}))#column as key and NA will replace with given value
################## ffill in fillna ##################
#The same interpolation methods available for reindexing can be used with fillna:
df2 = pd.DataFrame(np.random.randn(6, 3))
df2.iloc[2:, 1] = NA
df2.iloc[4:, 2] = NA
display(df2)
display(df2.fillna(method = 'ffill'))#it fill NA values with previous column values
display(df2.fillna(method='ffill', limit=2))#we can limit how many values filled inplace of NA from previous column
#With fillna you can do lots of other things with a little creativity. For example, you might pass the mean or median
#value of a Series:
data = pd.Series([1., NA, 3.5, NA, 7])
display(data)
display(data.fillna(data.mean()))#it fills NA values with the mean of whole data
```

### Removing Duplicates

```python
#The DataFrame method duplicated returns a boolean Series indicating whether each row is a duplicate 
#(has been observed in a previous row) or not:

dataa = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two']})
display(dataa.duplicated())# 0 and 1 row are not duplicated but 3 row is duplicated to 1 and so on so it gives TRUE
##Relatedly, drop_duplicates returns a DataFrame where the duplicated array is False:
display(dataa.drop_duplicates())

data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                     'k2': [1, 1, 2, 3, 3, 4, 4]}
                   )
display(data)

display(data.duplicated())#if any row duplicated to any other row in dataframe not necessary to its above row means it
#can be duplicated to any row present at any position in df it gives true when it comes second time in df
#in above case last row is completely duplicated of second last row (two,4) and (two,4)
############## actual ##############
display("duplicate value",data[data.duplicated()])
display(data.drop_duplicates())#remove last row of previous case
data1 = data
data1['v1'] = range(7)
display(data1)
display(data1.duplicated())#now duplicate
display(data.drop_duplicates(['k1']))#it only checks duplicated in rows of k1 excluding column k2 and v2
```

### Hierarchical Indexing

```python
#Hierarchical indexing is an important feature of pandas that enables you to have multiple (two or more) index levels on 
#an axis. Somewhat abstractly, it provides a way for you to work with higher dimensional data in a lower dimensional form

#################################### series ###########################################

data = pd.Series(np.random.randn(9),
                 index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],[1, 2, 3, 1, 3, 1, 2, 2, 3]])
display(data)
print(data.index)
#in above order of indexing matters like if 3 a in beginning and 1 a in the end then the series woulg show them in the 
#order you given to it.
dataa = pd.Series(np.random.randn(10),
                 index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd','a'],[1, 2, 3, 1, 3, 1, 2, 2, 3,1]])
display(dataa)
#With a hierarchically indexed object, so-called partial indexing is possible, enabling you to concisely select subsets 
#of the data:
display(data['a'])
display(data['a':'b'])
display(data.loc[['b', 'd']])
#Selection is even possible from an “inner” level:
display(data.loc[:, 2])
#Hierarchical indexing plays an important role in reshaping data and group-based operations like forming a pivot table. 
#For example, you could rearrange the data into a DataFrame using its unstack method:
display(data.unstack())#series dataframe where inner indexes serve as columns and outer indexes serves as a rows
#The inverse operation of unstack is stack:
display(data.unstack().stack())#dataframe to again series where columns becomes inner indexes and rows becomes outer.

################################### dataframe ##############################################

#With a DataFrame, either axis can have a hierarchical index:
frame = pd.DataFrame(np.arange(12).reshape((4, 3)),
                     index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]], 
                     columns=[['Ohio', 'Ohio', 'Colorado'],['Green', 'Red', 'Green']])
display(frame)
frame1 = frame
frame1.index.names = ['key1', 'key2']
frame1.columns.names = ['state', 'color']
display(frame1)
display(frame1.index)
display(frame1['Ohio'])
```

 ### Reordering and Sorting Level

```python
#The swaplevel takes two level numbers or names and returns a new object with the levels interchanged 
#(but the data is otherwise unaltered):
#it is to remember that level 0 is outside index and level 1 is inside index

frame = pd.DataFrame(np.arange(12).reshape((4, 3)),
                     index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]], 
                     columns=[['Ohio', 'Ohio', 'Colorado'],['Green', 'Red', 'Green']])
display(frame)
frame1 = frame
frame1.index.names = ['key1', 'key2']
frame1.columns.names = ['state', 'color']
display(frame1)
display(frame1.swaplevel(0,1))
#OR
display(frame1.swaplevel('key1','key2'))
#sort_index, on the other hand, sorts the data using only the values in a single level. When swapping levels, 
#it’s not uncommon to also use sort_index so that the result is lexicographically sorted by the indicated level
display(frame1.sort_index(level=0))#level = 0 means outer or key1 means a a then comes b b
display(frame1.sort_index(level=1))#level = 1 means inner or key2 means 1 1 then comes 2 2
#combine
display(frame.swaplevel('key1', 'key2').sort_index(level = 0))#first swap such that after swapping key2 = 0 and key1 = 1
#and after that sort index according to 0 which is key2 means 1 1 then comes 2 2
display(frame.swaplevel('key1', 'key2').sort_index(level = 1))#first swap such that after swapping key2 = 0 and key1 = 1
#and after that sort index according to 1 which is key1 means a a then comes b b
```

```python
#Merge or join operations combine datasets by linking rows using one or more keys.
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 
                    'data1': range(7)})
display(df1)
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
display(df2)
display(pd.merge(df1, df2))#since highes value till index 6 so we need 6 keys excluding keys not in common, pattern
#follows of df1 keys meeans b then a
#OR
display(pd.merge(df1, df2, on='key'))
display(pd.merge(df2, df1))#since highes value till index 6 so we need 6 keys excluding keys not in common, pattern
#follows of df2 keys meeans a then b
#if different column names
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
display(df3)
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
display(df4)
display(pd.merge(df3, df4, left_on='lkey', right_on='rkey'))#merge df3 with lkey,df4 with rkey,pattern of df3 as initial 
######################## outer join ##############################
display(pd.merge(df1, df2, how='outer'))
######################## inner join ##############################
#Many-to-many joins form the Cartesian product of the rows. Since there were three 'b' rows in the left DataFrame and 
#two in the right one, there are six 'b' rows in the result.
#3b*2b = 6b
#2a*2a = 4a
df5 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'data1': range(6)})
df6 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                    'data2': range(5)})
display(pd.merge(df5,df6, how='inner'))

#To merge with multiple keys, pass a list of column names:
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'], 
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
display(pd.merge(left, right, on=['key1', 'key2']))
display(pd.merge(left, right, on=['key1', 'key2'], how = 'outer'))
#In the case of combining Series along axis=1, the keys become the DataFrame column headers:
display(pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three']))
```

### Reshaping and Pivoting

```python
#stack This “rotates” or pivots from the columns in the data to the rows
#unstack This pivots from the rows into the columns
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
                    index=pd.Index(['Ohio', 'Colorado'], name='state'),
                    columns=pd.Index(['one', 'two', 'three'],name='number'))
display(data)
#Using the stack method on this data pivots the columns into the rows, producing a Series:
result = data.stack()
display(result)
display( result.unstack())
#By default the innermost level is unstacked (same with stack). You can unstack a different level by passing a 
#level number or name
#level = 0 -> outer
#level = 1 -> inner
display(result.unstack(0))
#OR
display(result.unstack('state'))
```
